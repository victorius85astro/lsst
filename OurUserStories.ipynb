```python
# This is a method that removes a fraction of points from a dataset at random with the same seed
def remove_fraction_with_seed(data, fraction, seed=0):
    """
    removes a fraction of data at random with given seed.
    :param data: data to remove
    :param fraction: fraction to remove
    :param seed: seed for randomness
    :return: data with fraction removed
    """
    n_to_remove = int(len(data) * fraction)
    np.random.seed(seed)

    return np.delete(data, np.random.choice(np.arange(len(data)), n_to_remove, replace=False))
```

```python
def corr2_coeff(A, B):
   
    # Rowwise mean of input arrays & subtract from input arrays themeselves
    A_mA = A - A.mean(1)[:, None]
    B_mB = B - B.mean(1)[:, None]

    # Sum of squares across rows
    ssA = (A_mA**2).sum(1)
    ssB = (B_mB**2).sum(1)

    # Finally get corr coeff np.dot(A_mA, B_mB.T) / np.sqrt(np.dot(ssA[:, None],ssB[None])); nije mi potrebno da delim
    #zbog velikih brojeva
    return (np.dot(A_mA, B_mB.T)) 
```

```python

###OVA FUNKCIJA TRAZI GRESKU PERIODA KAO 86 I 14 QUANTILE, ALI JE PROBLEM KOLIKO TACAKA U PEAK-U IMAMO, OVDE SE RUCNO UNOSI TREBA AUTOMATSKI
from traitlets.traitlets import Integer
def get_full_width(x: np.ndarray, y: np.ndarray, peak:np.ndarray,height: float = 0.5) -> float:
    height_half_max=y[peak[0]]*height
    index_max = peak[0]
    x_low = np.interp(height_half_max, y[index_max-9:index_max],  x[index_max-9:index_max])
    x_high = np.interp(height_half_max, np.flip(y[index_max:index_max+5]), np.flip(x[index_max:index_max+5]))
    arr=y[(x>=x_low)&(x<=x_high)]
    inversefunction = interpolate.interp1d(y[index_max-9:index_max],  x[index_max-9:index_max], kind='cubic')
    q3,q1 = np.quantile(arr, [0.86,0.14])
    er1=inversefunction(q1)
    er1 = np.interp(q1, y[:index_max], x[:index_max])
    er3 = np.interp(q3, y[:index_max], x[:index_max])
    return er1
```

```python

# TODO: Create bootstrap class with iid bootstrap, moving block bootstrap, write proper documentation and debug this! 
def block_bootstrap(mjd, mag,  block_length=10.0, rseed=None):
    np.random.seed(rseed) 
    N = len(mjd)
    mjd_boot = np.zeros(shape=(N, )) 
    mag_boot = np.zeros(shape=(N, ))

    k = 0
    last_time = 0.0
    for max_idx in range(2, N):
        if mjd[-1] - mjd[-max_idx] > block_length:
            break
    while k < N:
        idx_start = np.random.randint(N-max_idx-1) 
        for idx_end in range(idx_start+1, N):            
            if mjd[idx_end] - mjd[idx_start] > block_length or k + idx_end - idx_start >= N-1:
                break
        #print("%d %d %d %d" %(idx_start, idx_end, k, k + idx_end - idx_start))
        mjd_boot[k:k+idx_end-idx_start] = mjd[idx_start:idx_end] - mjd[idx_start] + last_time
        mag_boot[k:k+idx_end-idx_start] = mag[idx_start:idx_end]

        last_time = mjd[idx_end] - mjd[idx_start] + last_time
        k += idx_end - idx_start 
    return mjd_boot, mag_boot
```
